```python
#Build our least squares classifier for 2 classes
D = inputTrain.shape[1] + 1 #num of attributes, +1 is for the intercept (column of 1s)
K = expectedOutput.shape[1]
```

```python
expectedOutputTrain = np.asarray(expectedOutputTrain)
expectedOutputTest = np.asarray(expectedOutputTest)
inputTrain = np.asarray(inputTrain)
inputTest = np.asarray(inputTest)
validateTest = np.asarray(validateTest)
validateOutputTest = np.asarray(validateOutputTest)
```

```python
sum1 = 0
sum2 = 0
def createWeightMatrix(x, y, lmbda):
    numRows = x.shape[0]
    new_col = np.ones((numRows,1))
    augmentedX = np.c_[new_col, x]
    sum1 = np.dot(augmentedX.T, augmentedX) + lmbda
    sum2 = np.dot(augmentedX.T, y)
    W = np.dot(np.linalg.inv(sum1), sum2)
    return W
```

```python
def classificationTest(xTest, yTest, W, isValidating):  
    returnVals = []
    total = yTest.shape[0]
    predicted = np.zeros((total, K))
    i=0
    redSoilCorrect=cottonCropCorrect=greySoilCorrect=dampGreySoilCorrect=soilWithVSCorrect=veryDampGreySoilCorrect=totalCorrect=0
    correctPredictionCol = [0]*total
    for i in range(total):      
        x = xTest[i]        
        x = np.append(1,x)         
        x = x.reshape(1,D)        
        values = np.dot(W.T,x.T)
        values = values.T
        values = values.flatten()
        maxIndex = np.argmax(values)
        if yTest[i][maxIndex] == 1:
            if maxIndex == 0:
                redSoilCorrect += 1
            elif maxIndex == 1:
                cottonCropCorrect += 1
            elif maxIndex == 2:
                greySoilCorrect += 1
            elif maxIndex == 3:
                dampGreySoilCorrect +=1
            elif maxIndex == 4:
                soilWithVSCorrect +=1
            elif maxIndex == 5:
                veryDampGreySoilCorrect +=1
            correctPredictionCol[i] = 1
        predicted[i][maxIndex] = 1
    totalCorrect =redSoilCorrect+cottonCropCorrect+greySoilCorrect+dampGreySoilCorrect+soilWithVSCorrect+veryDampGreySoilCorrect
    print(totalCorrect)
    totalAccuracy=totalCorrect/float(total)*100
    if isValidating == False:
        redSoilAccuracy = redSoilCorrect/float(redSoilTotal)*100
        cottonCropAccuracy = cottonCropCorrect/float(cottonCropTotal)*100
        greySoilAccuracy = greySoilCorrect/float(greySoilTotal)*100
        dampGreySoilAccuracy = dampGreySoilCorrect/float(dampGreySoilTotal)*100
        soilWithVSAccuracy = soilWithVSCorrect/float(soilWithVSTotal)*100
        veryDampGreySoilAccuracy = veryDampGreySoilCorrect/float(veryDampGreySoilTotal)*100
        print("redSoil misclassification error = ", 100 - redSoilAccuracy)
        print("cottonCrop misclassification error = ", 100 - cottonCropAccuracy)
        print("greySoil misclassification error = ", 100 - greySoilAccuracy)
        print("dampGreySoil misclassification error = ", 100 - dampGreySoilAccuracy)
        print("soilWithVS misclassification error = ", 100 - soilWithVSAccuracy)
        print("veryDampGreySoil misclassification error = ", 100 - veryDampGreySoilAccuracy)
    else:
        redSoilAccuracy = redSoilCorrect/float(validationCounts[0])*100
        cottonCropAccuracy = cottonCropCorrect/float(validationCounts[1])*100
        greySoilAccuracy = greySoilCorrect/float(validationCounts[2])*100
        dampGreySoilAccuracy = dampGreySoilCorrect/float(validationCounts[3])*100
        soilWithVSAccuracy = soilWithVSCorrect/float(validationCounts[4])*100
        veryDampGreySoilAccuracy = veryDampGreySoilCorrect/float(validationCounts[5])*100
        print("redSoil misclassification error = ", 100 - redSoilAccuracy)
        print("cottonCrop misclassification error = ", 100 - cottonCropAccuracy)
        print("greySoil misclassification error = ", 100 - greySoilAccuracy)
        print("dampGreySoil misclassification error = ", 100 - dampGreySoilAccuracy)
        print("soilWithVS misclassification error = ", 100 - soilWithVSAccuracy)
        print("veryDampGreySoil misclassification error = ", 100 - veryDampGreySoilAccuracy)
    print("total misclassification error = ", 100 - totalAccuracy)
    print(total)
    returnVals.append(100-totalAccuracy)
    returnVals.append(predicted)
    returnVals.append(correctPredictionCol)
    return returnVals
```

```python
#cross validation
lambdaValues = [0, 100, 1000000,.1 , .3, .01, .03, .001, .003, .0001, .0003]
lambdaIndexOfMinError = -1
minError = 101
i = 0
for l in lambdaValues :
    print("lambda val?",l)
    W=0
    W = createWeightMatrix(inputTrain, expectedOutputTrain, l)
    print("W",W)
    error, predictedClasses, correctPredCol = classificationTest(validateTest, validateOutputTest, W, True)
    print("error ",error)    
    if(error < minError):
        minError = error
        lambdaIndexOfMinError = i
    i+=1
print("smallest index of error for lambda...", lambdaIndexOfMinError)

Wtest = createWeightMatrix(inputTrain, expectedOutputTrain, lambdaValues[lambdaIndexOfMinError])
```

```python
#testing
error, predictedClasses, correctPredictionCol = classificationTest(inputTest, expectedOutputTest, Wtest, False)
print(error)
```

```python
actual = pd.DataFrame(expectedOutputTest)
predictedDF = pd.DataFrame(predictedClasses)
correct = pd.DataFrame(correctPredictionCol)
frames = [predictedDF, actual, correct]
modelTest = pd.concat(frames, axis=1)
modelTest.columns = ["Predicted","Predicted", "Predicted", "Predicted", "Predicted", "Predicted", "Actual", "Actual", "Actual", "Actual", "Actual", "Actual", "Correct"]
modelTest
```
